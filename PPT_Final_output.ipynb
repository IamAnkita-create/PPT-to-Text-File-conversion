```
from pptx import Presentation
prs=Presentation('./Lorem17.pptx')
data= ''

for slides in prs.slides:
  for shapes in slides.shapes:
    if hasattr(shapes,"text"):
      print(shapes.text)
      data+=shapes.text
      
print(data)
file=open('extracted.txt','w')
file.write(data)
```

```
import os 
import nltk 
import nltk.corpus
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk import ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
import pandas as pd 
import numpy as np
```

```
h=[]
sentences = nltk.sent_tokenize(data)
h.append(sentences)
t_tokens=word_tokenize(data)
t_tag=nltk.pos_tag(t_tokens)
t_ner=ne_chunk(t_tag)
L=t_ner
#print(L)
print(h)
```

```
a=[]#contains noun
l1=[]#sentences of noun
l4=[]#tell if it's noun
for s in sentences:
  for word,pos in nltk.pos_tag(nltk.word_tokenize(str(s))):
    if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS'):
      a.append(word)
      l1.append(s)
      l4.append('noun')
print(l1)
```

```
b=[]#contains verbs
l2=[]#sentences of verb
l3=[]#tell if it's verb
for s in sentences:
  for ww,pos in nltk.pos_tag(nltk.word_tokenize(str(s))):
    if (pos == 'VB' or pos == 'VBD' or pos == 'VBG' or pos == 'VBN' or pos =='VBP' or pos =='VBZ'):
      b.append(ww)
      l2.append(s)
      l3.append('verb')
```

```
m1=l1+l2
for i in m1:
  print(i)
  #top=nouns
  #bottom=verbs
```

```
m2=a+b
for i in m2:
  print(i)
```

```
m3=l4+l3
for t in m3:
  print(t)
```
